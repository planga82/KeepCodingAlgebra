# Dimensión  

Definimos la **dimension** de un espacio vectorial V como el tamaño de una base para este espacio vectorial.  
La denotamos por dimV.  

Definimos el **rango** de un conjunto de vectores S como la dimensión de GenS.  
Lo escribimos como rangoS.

<span style="color:orange"> Ejemplo:  
Los vectores [1, 0, 0], [0, 2, 0], [2, 4, 0] son linealmente dependientes.  
Por tanto, su rango es menor que 3.  
Los dos primeros vectores forman una base para Gen{[1, 0, 0], [0, 2, 0], [2, 4, 0]}.  
Por tanto, el rango es 2.  
</style>  

<span style="color:orange"> Ejemplo:  
El espacio vectorial Gen{[0, 0, 0]} está generado por un conjunto "vacío" de vectores.  
Por lo tanto, el rango de {[0, 0, 0]} es 0.  
De hecho, es el único espacio vectorial con rango 0, todos los demás tienen al menos rango 1.
</style>  

Para una matriz M, el **rango por filas** de M es el rango de sus filas y el **rango por columnas** de M es el rango de sus columnas.  

<span style="color:orange"> Ejemplo:  
Sea la matriz $
  M=
  \left[ {\begin{array}{cc}
   1 & 0  & 0   \\
   0 & 2  & 0   \\
   2 & 4  & 0   \\
  \end{array} } \right]
$
Las filas de M son los vectores [1, 0, 0], [0, 2, 0], [2, 4, 0]. El conjunto formado por estos vectores {[1, 0, 0], [0, 2, 0], [2, 4, 0]} tiene rango dos, por lo que el rango por filas de M es 2.  
Las columnas de M son [1, 0, 2], [0, 2, 4], [0, 0, 0]. Como el tercer vector es el vector cero, no es necesario calcular el espacio vectorial generado por los tres vectores. Basta con observar que los dos primeros vectores tienen alguna coordenada 0 pero en diferentes posiciones, luego ambos son linalmente independientes. Por tanto, el rango por columnas es dos.  
¿Cómo lo podemos calcular en Python?</style>



<span style="color:orange"> Ejemplo:  
Sea la matriz $
  M=
  \left[ {\begin{array}{cc}
   1 & 0  & 0 & 5 \\

   0 & 0  & 3 & 9 \\
  \end{array} } \right]

Cada una de las filas tiene un valor distinto de cero donde las demás tienen ceros, por lo que las tres filas son linealmente independientes. Por lo tanto, el rango por filas de M es tres.

# Subbases  

Cada conjunto finito de vectores T contiene un subconjunto S que es una base para GenT.  

** Fin del algoritmo Grow**  
Si V es un subespacio de $\mathbb{F}$<sup>D</sup> donde D es finito, entonces el algoritmo Grow termina.  

**Teorema**  
Para D finito, cada subespacio de $\mathbb{F}$<sup>D</sup> contiene una base.  

**Lema**  
Sea V un espacio vectorial que consiste en D-vectores donde D es finito.  
Sea C un conjunto de vectores linealmente independientes que pertenecen a V.  
Entonces V tiene una base B que contiene todos los vectores de C.

# Suma directa de subespacios vectoriales

Si U y V comparten solo el vector **0**, entonces definimos la **suma directa** de U y V como el conjunto
{u + v: u∈U, v∈V}, denotado por U $\bigoplus$ V.  

Es decir, la suma directa es el conjunto de todas las sumas de un vector en U y un vector en V.  

<span style="color:orange"> Ejemplo:  
Sean U = Gen{1000, 0100} y V = Gen{0010}.</style>  
*<span style="color:orange">Cada vector distinto de cero en U tiene uno en la primera o segunda posición (o ambas) y en ninguna otra parte.</style>  
*<span style="color:orange">Cada vector distinto de cero en V tiene uno en la tercera posición y en ninguna otra parte.</style>  

<span style="color:orange">Por lo tanto, el único vector en U y V es el vector cero.
Por lo tanto:
U $\bigoplus$ V = {0000 + 0000, 1000 + 0000, 0100 + 0000, 1100 + 0000, 0000 + 0010, 1000 + 0010, 0100 + 0010, 1100 + 0010}={0000, 1000, 0100, 1100, 0010, 1010, 0110, 1110}.</style>  

<span style="color:orange">Ejemplo:  
Sea U = Gen{[1,2,1,2], [3,0,0,4]} y sea V=Null{[0,1,-1,0],[1,0,0,-1].  
El vector [2, -2, -1,2] está en U porque [2, -2, -1,2]=[3,0,0,4] - [1,2,1,2]
   También está en V porque $
\left[ {\begin{array}{cc}
   0 & 1  & -1 & 0 \\
   1 & 0  & 0 & -1 \\
  \end{array} } \right]
 \cdot \left[ {\begin{array}{cc}
   2 \\
   -2 \\
   -1 \\
   2 \\
  \end{array} } \right]=\left[ {\begin{array}{cc}
   0 \\
   0 \\
  \end{array} } \right]$.
Por lo tanto, no podemos formar U $\bigoplus$ V.</style>  

<span style="color:orange">Ejemplo:  
Sea U=Gen{[4,−1,1]} y V=Gen{[0,1,1]} sobre $\mathbb{R}$<sup>3</sup>.  
Son dos rectas cuya intersección está en el origen, luego puedo calcular U $\bigoplus$ V.  
U $\bigoplus$ V=Gen{[4,−1,1],[0,1,1]}</style>
